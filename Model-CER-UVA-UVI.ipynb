```python
#Convertir el tipo de datos de la columna indice_tiempo a datetime
df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'])
```

```python
import numpy as np
#Descarto las observaciones que no tienen uva y uvi
df = df.dropna()
#df = df[uva_uvi_filter]
df.head(5)
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
axis = plt.gca()

df.plot(kind='line', x='indice_tiempo', y='cer_diario', ax=axis)
df.plot(kind='line', x='indice_tiempo', y='uva_diario', ax=axis)
df.plot(kind='line', x='indice_tiempo', y='uvi_diario', ax=axis)
```

```python
#Detalle 3 ultimos aos
import datetime as dt
y_2017 = (df['indice_tiempo'] > pd.Timestamp(2017,1,1)) & (df['indice_tiempo'] <= pd.Timestamp(2017,12,31))
y_2018 = (df['indice_tiempo'] > pd.Timestamp(2018,1,1)) & (df['indice_tiempo'] <= pd.Timestamp(2018,12,31))
y_2019 = (df['indice_tiempo'] > pd.Timestamp(2019,1,1)) & (df['indice_tiempo'] <= pd.Timestamp(2019,12,31))

years = list()
years.append(y_2017)
years.append(y_2018)
years.append(y_2019)

for year in years:
    fig, axis = plt.subplots()
    
    df[year].plot(kind='line', x='indice_tiempo', y='cer_diario', ax=axis, figsize=(10,6))
    df[year].plot(kind='line', x='indice_tiempo', y='uva_diario', ax=axis)
    df[year].plot(kind='line', x='indice_tiempo', y='uvi_diario', ax=axis)

```
