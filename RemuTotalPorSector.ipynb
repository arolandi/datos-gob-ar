```python
import pandas as pd
df = pd.read_csv('rem-bruta-prom.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['indice_tiempo'] = pd.to_datetime(df['indice_tiempo'])
```

```python
df.info()
```

```python
df.head(5)
```

```python
import numpy as np
#Filtro por anio mayor a 2010
date_filter = df['indice_tiempo'] > '20100101'
df = df[date_filter]
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
axis = plt.gca()

df.plot(kind='line', x='indice_tiempo', y='intermediacion_financiera', ax=axis)
df.plot(kind='line', x='indice_tiempo', y='suministro_electricidad_gas_agua', ax=axis)
df.plot(kind='line', x='indice_tiempo', y='transporte_almacenamiento_comunicaciones', ax=axis)
df.plot(kind='line', x='indice_tiempo', y='explotacion_minas_canteras', ax=axis)

```

```python
#Filtro por anio mayor a 2015
date_flt = df['indice_tiempo'] > '20150101'

axis = plt.gca()

df[date_flt].plot(kind='line', x='indice_tiempo', y='intermediacion_financiera', ax=axis)
df[date_flt].plot(kind='line', x='indice_tiempo', y='suministro_electricidad_gas_agua', ax=axis)
df[date_flt].plot(kind='line', x='indice_tiempo', y='transporte_almacenamiento_comunicaciones', ax=axis)
df[date_flt].plot(kind='line', x='indice_tiempo', y='explotacion_minas_canteras', ax=axis)

fig, axis = plt.subplots()

df[date_flt].plot(kind='line', x='indice_tiempo', y='actividades_inmobiliarias_empresariales_alquiler', ax=axis)
df[date_flt].plot(kind='line', x='indice_tiempo', y='ensenianza', ax=axis)
df[date_flt].plot(kind='line', x='indice_tiempo', y='servicios_sociales_salud', ax=axis)
df[date_flt].plot(kind='line', x='indice_tiempo', y='otras_actividades_servicios_comunitarias_soc_personales', ax=axis)

```
